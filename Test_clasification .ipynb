```python
import csv
import pandas as pd 
```

```python
def solucion():
    #ESTA ES LA FUNCIÓN A LA QUE LE DEBES GARANTIZAR LOS RETORNOS REQUERIDOS EN EL ENUNCIADO.
    document = pd.read_csv('BTC-USD.csv')
    document = document.reindex(columns=['Date','Open','Close'])
    document['Variacion diaria'] = document['Close']-document['Open']
    document.loc[document['Variacion diaria'] > 0, 'Descripcion'] = 'Sube'
    document.loc[document['Variacion diaria'] < 0, 'Descripcion'] = 'Baja'
    document.loc[document['Variacion diaria'] == 0, 'Descripcion'] = 'Estable'
    document = document.rename(columns={'Date':'Fecha'}).reset_index()
    document = document.rename(columns={'index':'Indice'})
    mayor = document[['Close','Open']].max()
    mayor_precio = mayor.max()
    menor = document[['Close','Open']].min()
    menor_precio = menor.min()
    mayor = document.loc[document.Close == mayor_precio].reset_index()
    fecha_mayor_precio = mayor['Fecha'][0]
    menor = document.loc[document.Open == menor_precio].reset_index()
    fecha_menor_precio = menor['Fecha'][0]
    document.to_csv('analisis_bitcoin.csv', sep=';', index = False)
    variacion_diaria_media = document['Variacion diaria'].mean()
    return document#fecha_menor_precio, menor_precio, fecha_mayor_precio, mayor_precio, variacion_diaria_media

solucion()
```
